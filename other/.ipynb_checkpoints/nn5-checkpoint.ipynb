```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import random
```

```python
file_path="nn5.csv"
```

# Load data

```python
raw_data = []
with open(file_path) as f:
    lines = f.readlines()
    for line in lines:
        values = [v for v in line.split(',')]
        for i in range(len(values)):
            try:
                values[i] = float(values[i])
            except ValueError:
                values[i] = None
                continue
        if len(values) < 10:
            continue
        raw_data.append(values)

time_series_train = [[] for _ in range(len(raw_data[0]))]
time_series_test = [[] for _ in range(len(raw_data[0]))]
for j in range(len(raw_data)):
    nn5 = raw_data[j]
    for i in range(len(nn5)):
        if nn5[i] is None:
            continue
        if j < len(raw_data) - 56:
            time_series_train[i].append(nn5[i])
        else:
            time_series_test[i].append(nn5[i])

time_series_train = [np.array(v) for v in time_series_train]
time_series_test = [np.array(v) for v in time_series_test]
```

```python
def autocorr(time_series):
    diffs = []
    current = time_series[0]
    for i in range(1, len(time_series)):
        diffs.append(np.abs(time_series[i]-current))
        current = time_series[i]
    return diffs

diffs = [autocorr(ts) for ts in time_series_train]
mean_diffs = [np.mean(diff) for diff in diffs]
std_diffs = [np.std(diff) for diff in diffs]
plt.plot(mean_diffs)
plt.show()
plt.plot(std_diffs)
plt.show()
```

# Train lstm

```python
def split_sequence(time_series, in_size, out_size):
    x, y = [], []
    for i in range(len(time_series)):
        end_in = i+in_size
        start_out = i+in_size
        end_out = start_out+out_size
        if end_out >= len(time_series)-1:
            break
        seq_in = time_series[i:end_in]
        seq_out = time_series[start_out:end_out]
        x.append(seq_in)
        y.append(seq_out)
    return list(zip(x,y))
train_samples = []
IN_SIZE=56
OUT_SIZE=56
for ts in tqdm(time_series_train):
#    print(f"n size: {len(ts)}")
    train_samples += split_sequence(ts, IN_SIZE, OUT_SIZE)

chosen = train_samples
xs = np.array([np.array(x) for x,_ in chosen])
ys = np.array([np.array(y) for _,y in chosen])
#    return xs, ys
x_train=xs
y_train=ys
x_test = y_train[len(y_train)-1]
y_test = np.array([np.array(ts) for ts in time_series_test])
train_samples = (xs, ys)
print(f"Number train samples {len(train_samples)}")
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(IN_SIZE,1)),#, input_shape=(IN_SIZE,1)
#    tf.keras.layers.LSTM(256, activation='relu', return_sequences=True),
#    tf.keras.layers.LSTM(256, activation='relu', return_sequences=True),
    tf.keras.layers.LSTM(64, activation='relu'),
#    tf.keras.layers.Dense(128),
    tf.keras.layers.Dense(OUT_SIZE),
])
model.summary()
print("out:",model.output_shape)
model.compile(optimizer=tf.optimizers.Adam(),
    loss=tf.keras.losses.MeanAbsoluteError(),
    metrics=[tf.keras.metrics.MeanSquaredError()])
BATCH_SIZE=256
print(x_train.shape)
print(model.predict(x_train.reshape(-1,56,1)).shape)
print("x shape:",x_train.shape)
print("y shape:",y_train.shape)
print("Number batches: ", len(x_train) / BATCH_SIZE)
print("x test shape: ", x_test.shape)
print("y test shape: ", y_test.shape)

model.fit(x=x_train.reshape(-1,56,1),y=y_train,batch_size=BATCH_SIZE, verbose=1,epochs=100)
```
